```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def f1(x):
    return abs(x)

def f2(x):
    return x**2

def f3(x):
    return 2*x**2

def f4(x):
    return x**2 + 20
```

```python
population = [2,3,4]
functions = [f1, f2, f3, f4]

def proportional_selection():
    fitness_function_values = []
    selection_probabilities = []

    for f in functions:
        fitness_function_value = []
        selection_probability  = []
        for x in population:
            fitness_function_value.append(f(x))
        
        fitness_function_values.append(fitness_function_value)
        selection_probabilities.append(np.array(fitness_function_value)/ sum(fitness_function_value))

    return fitness_function_values, selection_probabilities


fitness_values, selection_probabilities = proportional_selection()
```

```python
df = pd.DataFrame(fitness_values, columns=population, index=['f1', 'f2', 'f3', 'f4'])
df
```

```python
df = pd.DataFrame(selection_probabilities, columns=population, index=['f1', 'f2', 'f3', 'f4'])
df
```

```python
for i, row in enumerate(selection_probabilities):
    fig, ax = plt.subplots()
    ax.set_title(f'f{i+1}')
    ax.pie(row, labels=population, autopct='%1.1f%%', startangle=90)
    fig.show()
```
